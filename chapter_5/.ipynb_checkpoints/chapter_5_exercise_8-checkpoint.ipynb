# Exercise 8
Train a LinearSVC on a linearly seperable dataset. Then train an SVC and a SGDClassifier on the same dataset. See if you can get them to produce roughly the same model

```python
%matplotlib inline
import itertools
import operator
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn import datasets
from sklearn.svm import LinearSVC, SVC
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import MinMaxScaler 
from tqdm import tqdm_notebook as tqdm
```

```python
RANDOM_STATE = 42
N_FEATURES = 4
N_INFORMATIVE = 3
N_REDUNDANT = 1
N_CLASSES = 3
N_CLUSTERS_PER_CLASS = 1
N_SAMPLES = 1000
TEST_SIZE = 0.20
```

```python
X, y = datasets.make_classification(n_features=N_FEATURES, n_classes=N_CLASSES, n_clusters_per_class=N_CLUSTERS_PER_CLASS,
                                    n_samples=N_SAMPLES, random_state=RANDOM_STATE, n_informative=N_INFORMATIVE, 
                                    n_redundant=N_REDUNDANT)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, 
                                                    random_state=RANDOM_STATE)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

```

```python
linearsvc_clf = LinearSVC(random_state=RANDOM_STATE)
linearsvc_clf.fit(X_train, y_train)
predictions = linearsvc_clf.predict(X_test)
cnf_mat = confusion_matrix(y_test, predictions)
plot_confusion_matrix(cnf_mat, classes=np.arange(N_CLASSES), normalize=True)
accuaracy = round(sum(np.diag(cnf_mat)) / sum(cnf_mat.flatten()), 2)
print('The accuracy is: {}'.format(accuaracy))
```

```python
svc_clf = SVC(random_state=RANDOM_STATE)
svc_clf.fit(X_train, y_train)
predictions = svc_clf.predict(X_test)
cnf_mat = confusion_matrix(y_test, predictions)
plot_confusion_matrix(cnf_mat, classes=np.arange(N_CLASSES), normalize=True)
accuaracy = round(sum(np.diag(cnf_mat)) / sum(cnf_mat.flatten()), 2)
print('The accuracy is: {}'.format(accuaracy))
```

```python
sgd_clf = SGDClassifier(random_state=RANDOM_STATE, max_iter=1000, tol=0.001)
sgd_clf.fit(X_train, y_train)
predictions = sgd_clf.predict(X_test)
cnf_mat = confusion_matrix(y_test, predictions)
plot_confusion_matrix(cnf_mat, classes=np.arange(N_CLASSES), normalize=True)
accuaracy = round(sum(np.diag(cnf_mat)) / sum(cnf_mat.flatten()), 2)
print('The accuracy is: {}'.format(accuaracy))
```

### Exercise 9
Train SVM classifier on MNIST dataset

```python
digits = datasets.load_digits()
X_train_mnist, X_test_mnist, y_train_mnist, y_test_mnist = train_test_split(digits.data, digits.target, 
                                                                            test_size=TEST_SIZE, random_state=RANDOM_STATE)
```

```python
labels = np.unique(digits.target)
models = dict()

param_dist = {
    'C': [0.9, 1.0, 1.1],
    'class_weight': [None, 'balanced'],
    'kernel': ['linear', 'rbf']
}

processed_images = []
for image in X_train_mnist:
    scaler = MinMaxScaler()
    image_processed = scaler.fit_transform(image.reshape(1, -1))
    processed_images.append(image_processed)

processed_data = np.array(processed_images).reshape(X_train_mnist.shape[0], X_train_mnist.shape[1])
for label in labels:
    new_train_labels = np.where(y_train_mnist == label, 1, 0)
    
    clf = SVC(probability=True, random_state=RANDOM_STATE)
    grid_search = GridSearchCV(clf, param_grid=param_dist, scoring='accuracy')
    grid_search.fit(X_train_mnist, new_train_labels)
    models[label] = grid_search
    
all_predictions = []
for true_label, data in tqdm(zip(y_test_mnist, X_test_mnist)):
    single_prediction = dict()
    
    for label in labels:
        model = models[label]
        positive_value = model.predict_proba(data.reshape(1, -1)).flatten()[1]
        single_prediction[label] = positive_value
    
    pred = max(single_prediction, key=single_prediction.get)
    all_predictions.append(pred)

cnf_mat = confusion_matrix(y_test_mnist, np.array(all_predictions))
plot_confusion_matrix(cnf_mat, classes=labels, normalize=True)
accuaracy = round(sum(np.diag(cnf_mat)) / sum(cnf_mat.flatten()), 2)
print('The accuracy is: {}'.format(accuaracy))
```
